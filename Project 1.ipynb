```python
# Import packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data

data = "World_countries_env_vars.csv"
df = pd.read_csv(data)

# Inspect data
df.head()
```

```python
# Create correlation matrix 

df_no_country_names = df.iloc[: , 1:]

correlations = df_no_country_names.corr()
correlations.head()
```

```python
# Try to isolate strong correlations (r > 0.59)

pd.set_option("display.max_rows", None, "display.max_columns", None)

strong_correlations = correlations[((correlations > 0.59) & (correlations < 1.0) | (correlations > -1.0) & (correlations < -0.59))]
strong_correlations.stack().reset_index(name = "Values")
```

```python
#Visualise the correlations

plt.figure(figsize=(17, 10))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.2)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Merge geometric information with environmental data

world_countries = world_countries.rename(columns={'ADMIN': 'Country'}) 
countries_and_data = pd.merge(df, world_countries, on="Country")

# Keep only rows where data is available on cropland cover

countries_and_data = countries_and_data[countries_and_data['cropland_cover'].notna()]

# Turn dataframe into a geodataframe (important for plotting)

gdf = geopandas.GeoDataFrame(countries_and_data, geometry=countries_and_data['geometry'])

# Create the map

gdf.plot(column = 'cropland_cover', legend = True)
```
